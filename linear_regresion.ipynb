# **Inciso 1: Estimación del Modelo de Regresión (Asumiendo que el intercepto es significativo)**

#### Estimar un modelo de regresión lineal simple mediante el método de los mínimos cuadrados. Elegir cuál es la variable dependiente (Y) y cuál es la independiente (X).

````python
import warnings
import statsmodels.api as sm
X = datos['Peso del Vehiculo (kg)'] # Variable independiente
Y = datos['Eficiencia Energetica (kWh/km)'] # Variable dependiente
X = sm.add_constant(X)  # se añade el intercepto
modelo = sm.OLS(Y, X).fit()
print(modelo.summary())
print("P-value: ", modelo.pvalues[1]) # P-value
print("R-squared: ", modelo.rsquared) # R-squared 
print("R-squared ajustado: ", modelo.rsquared_adj) 
print("F-statistic: ", modelo.fvalue) # F-statistic
#print("Correlacion: ", datos['Peso del Vehiculo (kg)'].corr(datos['Eficiencia Energetica (kWh/km)'])) # Correlacion con pandas
print("Correlacion: ", sp.stats.pearsonr(datos['Peso del Vehiculo (kg)'], datos['Eficiencia Energetica (kWh/km)'])) # Correlacion con scipy
warnings.filterwarnings('ignore')
````

````python
#grafico de regresion
import matplotlib.pyplot as plt

plt.scatter(datos['Peso del Vehiculo (kg)'], datos['Eficiencia Energetica (kWh/km)'])
plt.plot(datos['Peso del Vehiculo (kg)'], modelo.predict(X), color='red', label='Ajuste de mínimos cuadrados')
plt.xlabel('Peso del Vehiculo (kg)')
plt.ylabel('Eficiencia Energetica (kWh/km)')
plt.title('Relación entre Peso del Vehiculo y Eficiencia Energetica')
plt.legend()

# Mostrar la gráfica
plt.show()
````

````python
#por si acaso 
df= pd.array(modelo.resid)
print(len(df))
plt.hist(df, bins=20, color='c', edgecolor='k', alpha=0.7, label='Residuos')
plt.axvline(df.mean(), color='k', linestyle='dashed', linewidth=1)
plt.title('Histograma de los residuos')  
plt.show()
````

* `statsmodels.api as sm`: La librería `'statsmodels'` proporciona herramientas para la estimación de modelos estadísticos.
* `datos['Peso del Vehículo (kg)']`: columna correspondiente al peso del vehículo como la variable independiente (X).
* `datos['Eficiencia Energetica (kwh/km)']`: columna correspondiente a la eficiencia energética como la variable dependiente (Y).

````python
p_values = modelo.pvalues # extraer los valores p de los parametros
tvalue = modelo.tvalues # extraer los valores t de los parametros
t_pvalues = modelo.pvalues # extraer los valores p de los parametros
pdist = stats.t(modelo.df_resid) # distribucion t de los residuos
print("T-values: ", tvalue)
print("T-distribution p-values: ", t_pvalues)
print("T-distribution critical value (p=0.05): ", pdist.ppf(1-0.05))

for i in range(len(p_values)):
    print(f'P-valor de coeficiente β{i}: {round(p_values[i],5)}')
alpha = 0.05
significativos = p_values < alpha
for parametro, es_significativo in significativos.items():
    if es_significativo:
        print(f"El parámetro {parametro} es significativo (p < {alpha})")
    else:
        print(f"El parámetro {parametro} no es significativo (p >= {alpha})")
````
